```python
%matplotlib inline
```

```python
# Import Dependancies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis)
plt.xlabel('Months')
plt.ylabel('Fare ($)')
plt.grid()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis, color="green",marker="d",linewidth=2,label="Boston")
ax.legend()
ax.set_xlabel("Months")
ax.set_ylabel("Fare ($)")
ax.set_title("PyBer Fare by Month")
ax.grid()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Month")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the horizontal bar chart plot
plt.barh(x_axis, y_axis, color = "magenta", label = "Boston")
plt.gca().invert_yaxis()
plt.legend()
plt.ylabel("Month")
plt.xlabel("Fare ($)")
plt.title("PyBer Fare by Month")
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan",label="Chicago")
ax.invert_yaxis()
ax.set_ylabel('Months')
ax.set_xlabel('Fare ($)')
ax.set_title('PyBer Fare by Month')
```

```python
plt.plot(y_axis, x_axis, "o", color = "red", label = "Chicago")
plt.gca().invert_yaxis()
plt.legend()
plt.xlim(0,50)
plt.xlabel("Fare ($)")
plt.ylabel("Date")
plt.title("PyBer Fare by Month")
```

```python
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s = [i * 5 for i in y_axis], color = "skyblue", edgecolors = "black", alpha = .8, linewidths = 2,label = "Boston")
ax.invert_yaxis()
ax.set_xlabel("Fare ($)")
ax.set_ylabel("Date")
ax.legend()
ax.set_title("PyBer Fare by Month")
ax.set_xlim(0,50)
```

```python
plt.pie(y_axis, labels = x_axis)
plt.show()
```

```python
# Assign 12 colors, one for each month
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
        explode=explode_values,
        colors=colors,
        labels=x_axis, 
        autopct='%.1f%%',)
plt.show()
```

```python
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
fig, ax = plt.subplots(figsize=(8,8))
ax.pie(y_axis,
       labels=x_axis,
       colors=colors,
       explode=explode_values,
       shadow=True,
       startangle=95.0,
       counterclock=False,
       autopct='%1.1f%%')
plt.show()
```
